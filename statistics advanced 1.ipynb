The probability density function (PDF) of a continuous random variable is a function that describes the relative likelihood of the variable taking on a particular value. It is defined as the derivative of the cumulative distribution function (CDF).

The PDF is a non-negative function, and the total area under the PDF curve is equal to 1. This means that the probability of the variable taking on any value within a given range is equal to the area under the PDF curve within that range.

The PDF is a useful tool for understanding and modeling continuous random variables. It can be used to calculate the probability of a variable taking on a particular value, or the probability of it falling within a certain range. The PDF can also be used to generate random samples from a continuous distribution.




example of PDF

````python
import numpy as np
import matplotlib.pyplot as plt

# Define the PDF of a normal distribution with mean 0 and standard deviation 1
def normal_pdf(x):
  return np.exp(-0.5 * x**2) / np.sqrt(2 * np.pi)

# Generate a grid of x-values
x = np.linspace(-5, 5, 100)

# Calculate the PDF at each x-value
y = normal_pdf(x)

# Plot the PDF
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('PDF')
plt.title('Normal distribution PDF')
plt.show()
````

This plot shows the PDF of a normal distribution with mean 0 and standard deviation 1. The PDF is highest at the mean and decreases symmetrically away from the mean. The total area under the curve is equal to 1.

````python
import numpy as np

def normal_pdf(x, mean, std):
  """Calculates the probability density function of a normal distribution.

  Args:
    x: The value at which to calculate the PDF.
    mean: The mean of the normal distribution.
    std: The standard deviation of the normal distribution.

  Returns:
    The probability density function of the normal distribution at the given point.
  """

  return np.exp(-0.5 * ((x - mean) / std)**2) / (std * np.sqrt(2 * np.pi))
````

````python
import numpy as np

def poisson_cdf(k, lambda_):
  """Calculates the cumulative distribution function of a Poisson distribution.

  Args:
    k: The number of events.
    lambda_: The mean number of events.

  Returns:
    The cumulative distribution function of a Poisson distribution.
  """

  cdf = np.zeros(k + 1)
  cdf[0] = np.exp(-lambda_)
  for i in range(1, k + 1):
    cdf[i] = cdf[i - 1] + lambda_ * cdf[i - 1] / i
  return cdf

# Example usage:

lambda_ = 5
k = 10

cdf = poisson_cdf(k, lambda_)

print(cdf)
````

This function can be used to calculate the probability that a Poisson random variable will take on a value less than or equal to a certain value. For example, to calculate the probability that a Poisson random variable with mean 5 will take on a value less than or equal to 10, we would use the following code:



````python
import numpy as np

lambda_ = 5
k = 10
cdf = poisson_cdf(k, lambda_)
probability = cdf[10]
print(probability)
````

The binomial and Poisson distributions are two probability distributions that are used to model discrete events. However, there are some key differences between the two distributions.

**Binomial distribution**

The binomial distribution is used to model the probability of getting a certain number of successes in a fixed number of trials, where each trial has two possible outcomes (success or failure). The probability of success on each trial is constant and independent of the other trials.

**Poisson distribution**

The Poisson distribution is used to model the probability of getting a certain number of events in a given interval of time or space, where the events occur independently and at random. The mean number of events in the interval is constant.

Here is a table that summarizes the key differences between the binomial and Poisson distributions:

| Characteristic | Binomial distribution | Poisson distribution |
|---|---|---|
| Number of trials | Fixed | Infinite |
| Probability of success | Constant | Varies depending on the mean number of events |
| Independence of trials | Yes | Yes |
| Examples | Flipping a coin 10 times and counting the number of heads, rolling a die 6 times and counting the number of 6s | Counting the number of cars that pass through an intersection in 1 hour, counting the number of customers that visit a store in 1 day |

The Poisson distribution is typically used when the following conditions are met:
