Probability Mass Function (PMF)

A probability mass function (PMF) is a function that gives the probability that a discrete random variable will take on a certain value. The PMF must sum to 1 over all possible values of the random variable.

Example:

Consider a fair coin toss. The PMF for this random variable is as follows:

| Value | Probability |
|---|---|
| Heads | 0.5 |
| Tails | 0.5 |
This means that there is a 50% chance of getting heads on a coin toss, and a 50% chance of getting tails.

Probability Density Function (PDF)

A probability density function (PDF) is a function that gives the probability that a continuous random variable will take on a value within a certain interval. The PDF must integrate to 1 over the entire range of the random variable.

Example:

Consider a normal distribution with mean 0 and variance 1. The PDF for this random variable is as follows:

f(x) = exp(-x^2/2) / sqrt(2*pi)
This function gives the probability that the random variable will take on a value within a certain interval, such as the interval from -1 to 1. To calculate the probability that the random variable will take on a value within this interval, we would integrate the PDF over this interval:

P(-1 < x < 1) = int_{-1}^1 f(x) dx
This integral evaluates to 0.6827, which means that there is a 68.27% chance that the random variable will take on a value within the interval from -1 to 1.

Relationship between PMFs and PDFs

PMFs and PDFs are related to each other in that they both describe the probability distribution of a random variable. However, PMFs are used for discrete random variables, while PDFs are used for continuous random variables.

Another way to think about the difference between PMFs and PDFs is that PMFs give the probability that a random variable will take on a certain value, while PDFs give the probability that a random variable will take on a value within a certain interval.

Cumulative Density Function (CDF)

The cumulative distribution function (CDF) of a random variable is a function that gives the probability that the random variable will take on a value less than or equal to a certain value. It is defined as follows:

F(x) = P(X <= x)
where X is the random variable and F(x) is the CDF of X.

The CDF is a non-decreasing function, and it must approach 1 as x approaches infinity.

Example:

Consider the following CDF:

F(x) = 1 - exp(-x)
This is the CDF of the exponential distribution with parameter 1. The exponential distribution is a continuous distribution that is often used to model the time to failure of a component.

To calculate the probability that the random variable will take on a value less than or equal to a certain value, we can simply evaluate the CDF at that value. For example, the probability that the random variable will take on a value less than or equal to 1 is as follows:

P(X <= 1) = F(1) = 1 - exp(-1) = 0.6321

Formula for uniform distribution:

The probability density function (PDF) for the uniform distribution is as follows:

f(x) = 1 / (b - a)
where a and b are the lower and upper bounds of the distribution, respectively.

Example calculation:

Suppose we have a uniform distribution with lower bound 5 and upper bound 10. The probability that a randomly selected value from this distribution will be less than or equal to 6 is calculated as follows:

P(X <= 6) = (6 - 5) / (10 - 5) = 0.5
This makes sense because the probability of getting a value less than or equal to 6 is the same as the probability of getting a value greater than or equal to 6, since all outcomes are equally likely.

The z-score is a statistical measurement of how far a raw score is from the mean value of the normal distribution. It is calculated by subtracting the mean from the raw score and then dividing by the standard deviation.

**Formula for z-score:**

```
z = (x - μ) / σ
```

where:

* x is the raw score
* μ is the mean
* σ is the standard deviation

**Interpretation of z-scores:**

A z-score of 0 indicates that the raw score is equal to the mean. A z-score greater than 0 indicates that the raw score is greater than the mean, and a z-score less than 0 indicates that the raw score is less than the mean. The absolute value of the z-score indicates how many standard deviations the raw score is away from the mean.

**Importance of z-scores:**

Z-scores are important because they allow us to standardize data from different distributions. This means that we can compare data points from different distributions even if the distributions have different means and standard deviations.

Z-scores are also used in a variety of statistical tests, such as the z-test and the t-test. These tests can be used to compare two groups of data or to test hypotheses about a population.
